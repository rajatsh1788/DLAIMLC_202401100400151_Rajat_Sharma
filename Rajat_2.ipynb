```python
# ===== Perceptron from scratch =====

X = [75, 65, 80]
W = [4, 3, 5]
b = -1000
y = 1
lr = 0.01

for epoch in range(20):

    # Step 2
    z = sum(w*x for w, x in zip(W, X)) + b

    # Step 3
    y_pred = 1 if z >= 0 else 0

    # Step 4
    error = y - y_pred

    print(f"Epoch {epoch} | z={z:.2f} | pred={y_pred} | error={error}")

    if error == 0:
        break

    # Step 5
    for i in range(len(W)):
        W[i] += lr * error * X[i]

    # Step 6
    b += lr * error

print("\nFinal Weights:", W)
print("Final Bias:", b)
```
